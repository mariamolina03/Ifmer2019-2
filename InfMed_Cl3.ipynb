##Mesh - Descriptor

```
URL = "https://id.nlm.nih.gov/mesh/lookup/descriptor"
#Etiqueta que se quiere buscar
Descriptor = "Pain"
#Comparación de la etiqueta con la base de datos. Puede ser: 'exact', 'contains', 'startswith'
matched = "contains"
#Maximo número de registros
limit = 50
```

```
Parameters = {'label': Descriptor, 'match': matched, 'limit':limit}
```

```
respond = requests.get(url = URL, params = Parameters)
```

```
data = respond.json()
```
